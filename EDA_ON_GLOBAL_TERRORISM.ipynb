```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
```

```python
df= pd.read_csv("/content/drive/MyDrive/terrorism eda/terrorism-eda-dataset.csv", encoding = "ISO-8859-1")
df.head(10)
```

```python
df.shape
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','city':'City','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df['Damages'] = df.Killed + df.Wounded
df=df[['Year','Month','Day','Country','Region','City','latitude','longitude','AttackType','Killed','Wounded','Damages','Target','Group','Target_type','Weapon_type']]
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(15,6))
country_attack=df.Country.value_counts()[:15].reset_index()
country_attack.columns= ["Country", "Total Attacks"]
px.bar(data_frame= country_attack,x = 'Country',y = 'Total Attacks',color = 'Country',template='plotly_dark')
```

```python
plt.savefig('my-graph.png')
```

```python
plt.figure(figsize=(15, 8))
cas_count= df.groupby("Country").Damages.sum().to_frame().reset_index().sort_values("Damages", ascending=False)[:15]
px.bar(data_frame=cas_count,x = 'Country',y = 'Damages',color='Country',template='plotly_dark')
```

```python
plt.savefig('my-graph2.png')
```

FROM ABOVE TWO PLOTS OF TERROR ATTACKS VS COUNTRY AND DAMAGES VS COUNTRY, CLEARLY IRAQ HAS THE MOST TERROR ATTACKS AND DUE TO THIS AGAIN IRAQ HAS THE MOST DAMAGES.
